```python
import pandas as pd

train_data = pd.read_excel('Train.xlsx')
test_data = pd.read_excel('Test.xlsx')
train_data, test_data
```

```python
#Tiền xử lý dữ liệu
import regex as re
from nltk.corpus import stopwords

#Loại bỏ các từ dừng
def removeStopWords(document):
    words = [i for i in document.strip().split() if i not in stopwords.words('english')]
    return ' '.join(words)

def textPreprocess(document):
    # đưa về chữ viết thường
    document = document.lower()
    #loại bỏ ký tự đặc biệt
    document = re.sub('\W+',' ', document)
    #loại bỏ các từ dừng
    document = removeStopWords(document)
    #loại bỏ các siêu liên kết
    document = re.sub(r"http\S+", ' ', document)
    #loại bỏ các số
    document = ''.join([i for i in document if not i.isdigit()])
    # xóa khoảng trắng thừa
    document = re.sub(r'\s+', ' ', document).strip()
    return document

#tiền xử lý cho tất cả các cột của Data Frame
for i in train_data.columns:
    for j in range(len(train_data[i])):
        train_data[i][j] = textPreprocess(train_data[i][j])
for i in test_data.columns:
    for j in range(len(test_data[i])):
        test_data[i][j] = textPreprocess(test_data[i][j])

train_data, test_data
```

```python
#Hợp nhất 2 cột dữ liệu Subject và Text
train_data['Text'] =  train_data['Subject'] + train_data['Text']
train_data = train_data.drop(['Subject'], axis=1)
test_data['Text'] =  test_data['Subject'] + test_data['Text']
test_data = test_data.drop(['Subject'], axis=1)
train_data, test_data
```
