```python
#1. Hiển thị toàn bộ dữ liệu
import pandas as pd
import numpy as np

df1 = pd.read_csv('P4AI_BT1.csv')
df1
```

```python
#2. Điền các giá trị bị thiếu của dữ liệu theo phương pháp trung bình với các trường có giá trị số và theo phương pháp tần số cao nhất với giá trị rời rạc
df2 = df1.copy()
row_toal = df2.shape[0]
for i in df2.columns:
    if(i != 'variety'):
        sum = 0
        for j in df2[i].values:
            if(pd.isna(j) == False):
                sum += j       
        sum = round(sum / row_toal, 1)        
        df2[i] = df2[i].replace(np.nan, sum)
df2['variety'] = df2['variety'].replace(np.nan, list(dict(df2['variety'].value_counts()).keys())[0])
df2
```

```python
#3. Hiển thị các bản ghi có sepal.length > 5 và sepal.width > 3
df3 = df2.copy()
df3[(df3['sepal.length'].values > 5) & (df3['sepal.width'].values > 3)]
```

```python
#4. Chuẩn hóa dữ liệu theo min max
df4 = df3.copy()
for i in df4.columns:
    if(i != 'variety'):
        min_value = df4[i].min()
        max_value = df4[i].max()
        df4[i] = (df4[i] - min_value) / (max_value - min_value)
df4
```

```python
#5. Liên tục hóa thuộc tính variety
df5 = df4.copy()
dummies = pd.get_dummies(df5['variety'])
df5 = pd.concat([df5, dummies], axis=1)
df5.drop('variety', axis = 1, inplace=True)
df5
```

```python
#6. Xây dựng mô hình hồi quy tuyến tính để đoán kết quả của thuộc tính variety  dựa trên các thuộc tính khác
df6 = df4.copy().sample(frac = 1, replace=True) #xáo trộn dữ liệu để ngăn mô hình bị sai lệch
df6.reset_index(inplace = True)
df6.drop(['index'], axis=1, inplace=True)
df6['variety'].replace({'Setosa':1, 'Virginica':2, 'Versicolor':3}, inplace = True)
X = df6.drop('variety', axis=1)
Y = df6['variety']
thread_point = int(0.8 * df6.shape[0])
x_train, x_test = X[:thread_point], X[thread_point:]
y_train, y_test = Y[:thread_point], Y[thread_point:]

from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(x_train, y_train)
  
print(model.predict(x_test), y_test.values)
model.score(x_test, y_test)
```

```python
#7. Lấy mẫu ngẫu nhiên có lặp lại 50% dữ liệu
df7 = df5.copy()
df7 = df7.sample(frac=0.5, replace = False)
df7.reset_index(inplace = True)
df7.drop(['index'], axis=1, inplace=True)
df7
```

```python
#8. Rời rạc hóa thuộc tính sepal.length
df8 = df5.copy()
df8['sepal.length'] = pd.cut(df8['sepal.length'], bins=[0, 0.2, 0.4, 0.6, 0.8, 1], labels = ['very short', 'short', 'medium', 'long', 'very long'])
df8
```

```python
#9. Phân cụm dữ liệu và loại bỏ ngoại lai
df9a = df6.copy().set_index('variety')
df9a
```

```python
#Phân cụm dữ liệu
from sklearn.cluster import KMeans
clusters = 7

kmeans = KMeans(n_clusters = clusters)
kmeans.fit(df9a)
print(kmeans.inertia_)
kmeans.labels_
```

```python
#loại bỏ ngoại lai
class Pair:
    def __init__(self, value, cnt):
        self.value = value
        self.cnt = cnt    
        self.mine_value = 0

df9b = df6.copy()
columns = ['sepal.length', 'sepal.width', 'petal.length', 'petal.width']
dict = {}
for i in range(150):
    dict[i] = 0
#khởi tạo ngưỡng 
threshold_value = 0.2
#tính giá trị trung bình của từng cụm trên từng cột
for i in columns:
    d = {}
    for j in range(7):
        d[j] = Pair(0, 0)
    #cộng giá trị của từng cụm và tần suất xuất hiện
    for j in range(len(kmeans.labels_)):
        d[kmeans.labels_[j]].value += df9b[i][j]
        d[kmeans.labels_[j]].cnt += 1
    #tính giá trị trung bình
    for j in range(7):
        d[j].mine_value = d[j].value / d[j].cnt  
    #đếm số cột có chênh lệch với giá trị trung bình vượt ngưỡng của từng dòng
    for j in range(150):
        if(abs(d[kmeans.labels_[j]].mine_value - df9b[i][j]) > threshold_value):
            dict[j] += 1
dict
```

```python
#hiển thị tần suất của số cột có chênh lệch vượt ngưỡng trên từng dòng
for i in range(5):
    print('{} : {}'.format(i, list(dict.values()).count(i))) 
```

```python
# những dòng có tần suất lớn hơn 2 là ngoại lai và sẽ bị loại bỏ
for i in range(150):
    if (dict[i] > 0):
        df9b.drop(i, inplace=True)
df9b.reset_index(inplace = True)
df9b.drop(['index'], axis=1, inplace=True)        
df9b
```

```python
#training lại sau khi loại bỏ ngoại lai
X = df9b.drop('variety', axis=1)
Y = df9b['variety']
thread_point = int(0.8 * df9b.shape[0])
x_train, x_test = X[:thread_point], X[thread_point:]
y_train, y_test = Y[:thread_point], Y[thread_point:]

model = LogisticRegression()
model.fit(x_train, y_train)
  
print(model.predict(x_test), y_test.values)
model.score(x_test, y_test) #sau khi loại bỏ ngoại lai tỉ lệ dự đoán tăng từ 0.9 lên 0.92
```
